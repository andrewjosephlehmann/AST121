```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
```

```python
### Redshift as a function of your radius to the center of a blackhole ###

def zschwarzchild(z):
    return (1/(1-(1/((z+1)**2))))

#Verify it's function when z = 4# 

print(zschwarzchild(0.001))
```

```python
def isco(r,x):
    return r + 0*x
print(isco(3,1))
```

```python
zlin = np.linspace(0.05, 10, 1000)
```

```python
zsch = zschwarzchild(zlin)
```

```python
min_isco = isco(1.5,zlin)
rot_min_isco = isco(4.5, zlin)
```

```python
# Redshift #

#Here is where we use matplotlib to graph our function
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
#ax.spines['left'].set_position('left')
#ax.spines['bottom'].set_position('b')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

# plot the functions
plt.plot(zlin, zsch, label='Redshift as function of radii')
plt.plot(zlin, min_isco, label='Photon Sphere')
plt.plot(zlin, rot_min_isco)


### Labels ###
plt.xlabel('Redshift (z)')
plt.ylabel('$r/R_s$')
plt.title('Redshift as a function of radius')
plt.legend(loc='upper right')


# show the plot
plt.show()
```

```python
### The speed at which Quazars are moving away from us ###

def zlongspeed(v):
    return ((np.sqrt((1 + v)/(1 - v))) - 1)

print(zlongspeed(0.1))
```

```python
zspeed = np.linspace(0,1, 100)
```

```python
### function that demonstrates recession velocity in relation to redshift ###

zshift = zlongspeed(zspeed)
```

```python
# Redshift # 

#Here is where we use matplotlib to graph our function
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
#ax.spines['left'].set_position('left') 
#ax.spines['bottom'].set_position('b')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

# plot the functions
plt.plot(zshift,zspeed )



### Labels ###
plt.xlabel('Redshift (z)')
plt.ylabel('Recession Velocity (c)')
plt.title('Redshift in relation to Recession Velocity')
#plt.legend(loc='upper left')


# show the plot
plt.show()
```

```python
#HUBBLE TO AGE# 
def hubbleage(h):
    return (((h/(3.086*10**19))**(-1))*(1/(((5**5 * 4**4 * 3**3 * 2**2 * 1**1)*(10**-3))*(365)))*(10**(-9)))
```

```python
print(hubbleage(71.265))
```

```python
def uniage(a):
    return (((hubbleage(71.265)**(2/3))*(1/(1+a)))**(3/2))
```

```python
print(uniage(100))
```

```python
fig, (ax1) = plt.subplots(1, 1, sharex=True, sharey=True)
fig.suptitle('Wave Functions')


line1, = ax1.plot([], [], color='r', lw=2)
line2, = ax1.plot([], [], color='g', lw=2)
line3, = ax1.plot([], [], color='b', lw=2)
```

```python
def init():
        line1.set_data([], [])
        line2.set_data([], [])
        line3.set_data([], [])
        return line1, line2, line3
```

```python
def animate(i):
        x = np.linspace(0,10,100000)
        y1 = uniage(i)
        line1.set_data(x, y1)
```

```python
anim2 = FuncAnimation(fig, animate, init_func=init,
                     frames=100, interval=100, blit=True)
```
