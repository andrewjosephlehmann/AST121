```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
### Calculating relativistic doppler shift ###

def reldop (beta,wavelengthi):
    return wavelengthi*(np.sqrt((1+beta)/(1-beta)))

print(reldop(-0.9945,1*10**-2))
```

```python
x = np.linspace(-1,1, 100)
y = reldop(x,1)
```

```python
#Here is where we use matplotlib to graph our function
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
#ax.spines['left'].set_position('left')
#ax.spines['bottom'].set_position('b')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

# plot the functions
plt.plot(x,y)

### Labels ###
plt.xlabel('Values of Beta')
plt.ylabel('$\delta\lambda$')
plt.title('Relativistc Doppler')
# plt.legend(loc='upper left')

# show the plot
plt.show()
```

```python
### Non-Relativistic Doppler ###

def dop (wavelengthi2,beta2):
    return (beta2*wavelengthi2+wavelengthi2)

print(dop(656.3,0.00067))
```

```python
y2 = dop(1, x)
```

```python
#Here is where we use matplotlib to graph our function
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
#ax.spines['left'].set_position('left')
#ax.spines['bottom'].set_position('b')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

# plot the functions
plt.plot(x,y)
plt.plot(x,y2)

### Labels ###
plt.xlabel('Values of Beta')
plt.ylabel('$\delta\lambda$')
plt.title('Relativistic Doppler vs Classical Doppler')
plt.plot(x,y, label='Relativistic Doppler')
plt.plot(x,y2, label='Classical Doppler')
plt.legend(loc='upper left')


# show the plot
plt.show()
```

```python
### Redshift in function of recession velocity ###

def redv(z):
    return ((z*(z+2)) / (z*(z+2)+2))
```
