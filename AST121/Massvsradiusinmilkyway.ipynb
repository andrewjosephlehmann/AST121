```python
import numpy as np
import matplotlib.pyplot as plt
import csv
```

```python
### Data points ###

### Radius in kpc ###

rdata = np.array([2.9, 4.2, 5.5, 6.5, 7.4, 8.0, 8.5])

### Velocity in km/s ###

vdata = np.array([205, 229, 237, 238, 252, 237, 230])
```

```python
#Here is where we use matplotlib to graph our function
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
#ax.spines['left'].set_position('left')
#ax.spines['bottom'].set_position('b')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

# plot the functions
plt.plot(rdata,vdata)

### Labels ###
plt.xlabel('Radius (kpc)')
plt.ylabel('Velocity (km/s)')
plt.title('Velocity vs Radius of Milky Way')
# plt.legend(loc='upper left')

# show the plot
plt.show()
```

```python
### Mass density function ###

def Massdensity(velocity, radius):
    return (velocity**2)*radius/6.67*10**-11

print(Massdensity(1,2))


### Total mass should be 10^10 for total density of milky way 
```

```python
#Convert Kpc to meters# 

def KpcM(lengthinkpc):
    return (3.086*10**19)*(lengthinkpc)

#Verify if the function, works nominally#
print(KpcM(1.2))
```

```python
#Convert km/s to m/s#

def kmstoms(kms):
    return (1000)*(kms)
#Verify if the function is functioning nominally#
print(kmstoms(1))
```

```python
msvdata = kmstoms(vdata)

print(msvdata)
```

```python
rmdata = KpcM(rdata)

print(rmdata)
```

```python
# Mass Density of Milky Way #

MdenMW = Massdensity(msvdata, rmdata)
```

```python
#Here is where we use matplotlib to graph our function
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
#ax.spines['left'].set_position('left')
#ax.spines['bottom'].set_position('b')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

# plot the functions
plt.plot(rdata,MdenMW)

### Labels ###
plt.xlabel('Radius (kpc)')
plt.ylabel('Mass Density (kg)')
plt.title('Mass Distribution of Milky Way')

# plt.legend(loc='upper left')

# show the plot
plt.show()
```
